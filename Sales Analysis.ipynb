```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Sales = pd.read_csv('Sales-Python.csv')
```

```python
Sales.head(10)
```

### Descriptive statistics

```python
Sales.describe()
```

```python
Sales[['Row_ID','Order_ID','Order_Quantity','Sales','Discount','Profit','Unit_Price','Shipping_Cost','Product_Base_Margin']].corr()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(Sales["Sales"], kde=True, bins=30, color="skyblue")
plt.title("Sales Distribution", fontsize=16)
plt.xlabel("Sales", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
```

#### Profit vs. Discount (Relationship)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Sales, x="Discount", y="Profit", hue="Region", alpha=0.7, palette="tab10")
plt.title("Profit vs. Discount by Region", fontsize=16)
plt.xlabel("Discount", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.legend(title="Region")
```

#### Sales by Region

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=Sales, x="Region", y="Sales", palette="Set3")
plt.title("Sales by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Sales", fontsize=12)
```

#### Profit Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(Sales["Profit"], kde=True, bins=30, color="salmon")
plt.title("Profit Distribution", fontsize=16)
plt.xlabel("Profit", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

#### Sales and Profit by Product Category

```python
plt.figure(figsize=(10, 6))
category_sales_profit = Sales.groupby("Product_Category")[["Sales", "Profit"]].sum().reset_index()
category_sales_profit = category_sales_profit.melt(id_vars="Product_Category", var_name="Metric", value_name="Value")

sns.barplot(data=category_sales_profit, x="Product_Category", y="Value", hue="Metric", palette="viridis")
plt.title("Sales and Profit by Product Category", fontsize=16)
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Total Value", fontsize=12)
plt.legend(title="Metric")
```

#### Outliers in Profit

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=Sales, x="Region", y="Profit", palette="coolwarm")
plt.title("Profit Distribution by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Profit", fontsize=12)
```
